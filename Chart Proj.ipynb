```python
lakersdf=df[df['fran_id']=='Lakers']
```

```python
pistonssdf=df[df['fran_id']=='Pistons']
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(lakersdf['year_id'],lakersdf['forecast'], color='purple')
plt.ylabel('Forecasted Winning Percentages')
plt.title('Lakers Winning Chances')

plt.subplot(1, 2, 2)
plt.plot(pistonssdf['year_id'], pistonssdf['forecast'], color='blue')
plt.ylabel('Forecasted Winning Percentages')
plt.title('Lakers Winning Chances')
plt.show()
```

The plots above represent fivethirtyeight's forcasted winning percentage for two of the NBA's historic franchises, the Lakers franchise in purple on the left and the Pistons in blue on the right. One could see from the fact that there are much fewer troughs in the Lakers data is that the Lakers are the more succesful franchise over their history. One could also look through each of these charts themselves and see the basic "success timeline" for each one of these franchises. O
